```python
msft_sk.Close.plot(title="closing price")
msft_sk['20_days']= msft_sk['Close'].rolling(window=20).mean()
msft_sk['50_days']= msft_sk['Close'].rolling(window=50).mean()
msft_sk['100_days']= msft_sk['Close'].rolling(window=100).mean()
plt.plot(msft_sk['20_days'], label= '20-days MA')
plt.plot(msft_sk['50_days'], label= '50-days MA')
plt.plot(msft_sk['100_days'], label= '100-days MA')
plt.legend()
```

```python
# daily Returns
msft_sk['Return']=msft_sk['Close'].pct_change()
```
